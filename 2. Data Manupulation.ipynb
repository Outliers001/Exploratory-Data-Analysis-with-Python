```python
marketing_data = marketing_data[['ID', 'Year_Birth', 'Education', 'Marital_Status',
                'Income', 'Kidhome', 'Teenhome', 'Dt_Customer', 
                'Recency','NumStorePurchases', 'NumWebVisitsMonth']]
```

#### 1.Group Data

```python
# Group by 'Education'
grouped_data = marketing_data.groupby('Education').mean()
# Group by 'Education' and calculate the mean of 'Income'
grouped_data_income = marketing_data.groupby('Education')['Income'].mean()

# Group by 'Marital_Status' and calculate the sum of 'NumStorePurchases'
grouped_purchases = marketing_data.groupby('Marital_Status')['NumStorePurchases'].sum()

print(grouped_data)
print(grouped_data_income)
print(grouped_purchases)
```

#### 2.Sorting Data:
To sort data by one or more columns:

```python
# Sort by 'Income'
sorted_data = marketing_data.sort_values(by='Income', ascending=False)

print(sorted_data)
```

#### 3. Categorizing Data:
You can convert continuous data into categories or bins.

```python
# Categorizing 'Income' into bins
marketing_data['Income_Category'] = pd.cut(marketing_data['Income'], 
                                           bins=[0, 30000, 60000, 90000, 120000], 
                                           labels=['Low', 'Medium', 'High', 'Very High'])

```

```python
marketing_data[['Income', 'Income_Category']]
```

#### 4. Changing Data Format:
You can change the format of data, for instance, converting date strings into a datetime format.


```python
# Convert 'Dt_Customer' to datetime format
marketing_data['Dt_Customer'] = pd.to_datetime(marketing_data['Dt_Customer'], format='%d-%m-%Y')

marketing_data['Dt_Customer']
```

#### 5. Replacing Data:
You can replace values in the dataset using .replace().

```python
# Finding the unique categories in 'Marital_Status'
unique_marital_status = marketing_data['Marital_Status'].unique()

print("Unique categories in 'Marital_Status':", unique_marital_status)
```

```python
# Replacing 'Marital_Status' categories
marketing_data['Marital_Status_Replace'] = marketing_data['Marital_Status'].replace({
    'Married': 'M', 
    'Single': 'S', 
    'Together': 'T',
    'Widow' : 'W',
    'Divorced' : 'D',
    'Alone' : 'A',
    'Absurd' : 'Ab',
    'YOLO' : 'YL'
})

print(marketing_data['Marital_Status_Replace'])
```

```python
marketing_data[['Marital_Status','Marital_Status_Replace']]
```

```python
marketing_data['Marital_Status_Replace_1'] = marketing_data['Marital_Status'].replace(['Single', 'Together', 'Married', 'Divorced', 'Widow', 'Alone', 'Absurd', 'YOLO'],['S','T','M','D','W','A','Ab','YL'])
```

```python
marketing_data[['Marital_Status','Marital_Status_Replace','Marital_Status_Replace_1']]
```

##### 10.Concatenating data
Sometimes, we may need to stitch multiple datasets or samples of the same dataset by columns and
not rows. This is where we concatenate our data. While appending stitches rows of data together,
concatenating stitches columns together to provide a single dataset. For example, if we have 2
datasets containing 1,000 rows and 20 columns each, the concatenated data will contain 1,000 rows
and 40 columns. The columns typically increase while the rows remain the same. The datasets are
allowed to have a different number of columns but typically should have the same number of rows to
avoid errors after concatenating.
In pandas, the concat method helps us concatenate data

```python
marketing_sample1 = pd.read_csv("data/marketing_campaign_concat1.csv")
marketing_sample2 = pd.read_csv("data/marketing_campaign_concat2.csv")
```

```python
marketing_sample1.shape
```

```python
marketing_sample1.head()
```

```python
marketing_sample2
```

```python
concatenated_data = pd.concat([marketing_sample1, marketing_sample2], axis = 1)
```

#### 11. Duplicates

```python
marketing_data = marketing_data[['Education', 'Marital_Status',
                'Kidhome', 'Teenhome']]
```

```python
# Find duplicate rows
duplicates = marketing_data.duplicated()

duplicates.sum()
```

```python
# Display duplicate rows
duplicate_rows = marketing_data[marketing_data.duplicated()]
duplicate_rows
```

```python
# Remove duplicate rows
marketing_data_no_duplicates = marketing_data.drop_duplicates()

marketing_data_no_duplicates
```

```python
# Find duplicates based on the 'ID' column
duplicates_by_id = marketing_data.duplicated(subset=['ID'])

print(duplicates_by_id)

# Remove duplicates based on 'ID' column
marketing_data_no_duplicates = marketing_data.drop_duplicates(subset=['ID'])

print(marketing_data_no_duplicates)
```
